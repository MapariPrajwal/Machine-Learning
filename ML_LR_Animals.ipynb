```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv("/content/mammals.csv")
data.replace('?', pd.NA, inplace=True)
X = data[['body_wt', 'non_dreaming', 'dreaming', 'total_sleep', 'life_span', 'gestation', 'predation', 'exposure', 'danger']]
y = data['brain_wt']  # Let's use 'brain_wt' as the target variable for demonstration purposes
numerical_cols = X.columns

# Impute missing values for numerical columns with mean
numerical_imputer = SimpleImputer(strategy='mean')
X[numerical_cols] = numerical_imputer.fit_transform(X[numerical_cols])

# Split the dataset into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
model = LinearRegression()
model.fit(X_train, y_train)
accuracy = model.score(X_test, y_test)
print("Accuracy Score:", accuracy)
y_pred = model.predict(X_test)

# Plot linear model
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred)
plt.xlabel("True Values")
plt.ylabel("Predictions")
plt.title("True Values vs. Predictions")
plt.show()
```
